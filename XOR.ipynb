```python
import numpy as np

from network import Network
from layers import Dense, ActivationLayer
from activations import tanh,sigmoid,tanh_prime,sigmoid_prime
from loss import mse, mse_prime

```

```python
# training data
x_train = np.array([[[0,0]],[[0,1]],[[1,0]],[[1,1]]])
y_train = np.array([[[0]],[[1]],[[1]],[[0]]])
# x_train.shape (4, 1, 2)
```

```python
# network
net = Network()
net.add(Dense(2,3,'FC_1'))
# net.add(ActivationLayer(tanh, tanh_prime, 'activ_1'))
net.add(ActivationLayer(tanh, tanh_prime, 'activ_1'))
net.add(Dense(3,1,'FC_2'))
net.add(ActivationLayer(tanh, tanh_prime, 'activ_2'))
net.use_loss(mse,mse_prime)
```

```python
# Train
net.fit(x_train, y_train, epochs=1000, learning_rate=0.1)
```

```python

# Test
print('testing with training data')
out = net.predict(x_train)
print(out)
```
